```python
# Create some sample data
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create the plot
plt.figure(figsize=(8, 4)) # Optional: set the figure size
plt.plot(x, y_sin, label='Sine wave')
plt.plot(x, y_cos, label='Cosine wave', linestyle='--')

# Add labels, title, and legend
plt.xlabel('X value')
plt.ylabel('Y value')
plt.title('Simple Sine and Cosine Plot')
plt.legend()
```

```python
arr = np.array([[1.0, 2.0, 3.0],
                [4.0, 8.0, 6.0],
                [7.0, 8.0, 9.0]])

diag = np.diag(arr)
exp_diag = np.exp(diag)
softmax_diag = exp_diag / np.sum(exp_diag)
plt.plot(diag,softmax_diag,marker = "o",color = "g",lw = 0.7)
plt.grid(True)
plt.title("soft vs diag")
plt.xlabel("soft")
plt.ylabel("diag")
```
